```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

calorie_intake = [2000, 2200, 2500, 1800, 2300, 2600, 2100, 1900, 2700, 3000]
exercise_duration = [60, 50, 30, 90, 45, 20, 75, 95, 15, 10]
bmi = [25, 26, 28, 23, 27, 30, 24, 22, 31, 33]

plt.scatter(calorie_intake, bmi)
plt.xlabel('Daily Calorie Intake')
plt.ylabel('BMI')
plt.title('Daily Calorie Intake vs. BMI')
plt.show()

plt.scatter(exercise_duration, bmi)
plt.xlabel('Exercise Duration')
plt.ylabel('BMI')
plt.title('Exercise Duration vs. BMI')
plt.show()

plt.scatter(calorie_intake, exercise_duration)
plt.xlabel('Daily Calorie Intake')
plt.ylabel('Exercise Duration')
plt.title('Daily Calorie Intake vs. Exercise Duration')
plt.show()

corr_calorie_bmi, _ = pearsonr(calorie_intake, bmi)
corr_exercise_bmi, _ = pearsonr(exercise_duration, bmi)
corr_calorie_exercise, _ = pearsonr(calorie_intake, exercise_duration)

print(f"Pearson Correlation (Calorie Intake vs. BMI): {corr_calorie_bmi:.3f}")
print(f"Pearson Correlation (Exercise Duration vs. BMI): {corr_exercise_bmi:.3f}")
print(f"Pearson Correlation (Calorie Intake vs. Exercise Duration): {corr_calorie_exercise:.3f}")

print("\nInterpretation:")
print(f"Calorie Intake vs. BMI ({corr_calorie_bmi:.3f}): Positive correlation indicates higher calorie intake is associated with higher BMI.")
print(f"Exercise Duration vs. BMI ({corr_exercise_bmi:.3f}): Negative correlation suggests longer exercise duration is associated with lower BMI.")
print(f"Calorie Intake vs. Exercise Duration ({corr_calorie_exercise:.3f}): Negative correlation implies higher calorie intake is linked to shorter exercise duration.")
print("\nDiscussion:")
print("Consuming more calories appears to lead to a higher BMI based on the positive correlation.")
print("Longer exercise duration tends to reduce BMI, as shown by the negative correlation.")
```
