Ques on 1
A customer service center receives calls daily, with the number of calls ranging from 0 to 7 per day. The management wants to study the distribu on of daily calls.
•	Generate a dataset represen ng the number of calls received per day for N days.
•	Compute and visualize the PMF and CDF to analyze the probability of receiving a specific number of calls on a given day.


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter


np.random.seed(42)


N = 1000
calls_per_day = np.random.randint(0, 8, size=N)


call_counts = Counter(calls_per_day)
pmf = {k: v / N for k, v in sorted(call_counts.items())}


sorted_calls = sorted(pmf.keys())
cdf = np.cumsum([pmf[k] for k in sorted_calls])

fig, ax = plt.subplots(1, 2, figsize=(12, 5))


sns.barplot(x=list(pmf.keys()), y=list(pmf.values()), ax=ax[0], palette="viridis")
ax[0].set_title("Probability Mass Function (PMF)")
ax[0].set_xlabel("Number of Calls per Day")
ax[0].set_ylabel("Probability")
ax[0].set_xticks(sorted_calls)


sns.lineplot(x=sorted_calls, y=cdf, marker="o", ax=ax[1], color="b")
ax[1].set_title("Cumulative Distribution Function (CDF)")
ax[1].set_xlabel("Number of Calls per Day")
ax[1].set_ylabel("Cumulative Probability")
ax[1].set_xticks(sorted_calls)

plt.tight_layout()
plt.show()
```

Ques on 2
A hospital receives emergency cases daily, and there is a 25% probability that a randomly arriving pa ent requires immediate surgery. Suppose N pa ents visit the emergency department in a day (N ≥ 100):
i) 	What 	is 	the 	probability 	that 	exactly 	20 	pa ents 	need 	surgery? ii) 	What 	is 	the 	probability 	that 	at 	least 	30 	pa ents 	need 	surgery? iii) On average, how many pa ents requiring surgery should the hospital expect?


```python
from scipy.stats import binom

N = 100
p = 0.25

k1 = 20
prob_20 = binom.pmf(k1, N, p)

k2 = 30
prob_at_least_30 = 1 - binom.cdf(k2 - 1, N, p)

expected_patients = N * p

print(f"1. Probability that exactly 20 patients need surgery: {prob_20:.4f}")
print(f"2. Probability that at least 30 patients need surgery: {prob_at_least_30:.4f}")
print(f"3. Expected number of patients requiring surgery per day: {expected_patients}")
```
